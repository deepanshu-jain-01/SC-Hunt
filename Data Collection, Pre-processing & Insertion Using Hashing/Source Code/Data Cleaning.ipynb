### Working on Job Type

```python
jobtype=df['job_type'].str.split(',')
df['job_type']=jobtype.str[0]
df['job_type'][df['job_type']=='Full Time Employee']='Full Time'
df['job_type'][df['job_type']=='Part Time Employee']='Part Time'
```

```python
df['job_type'] = df['job_type'].replace({
    'Full Time  Employee':'Full Time',
    'Employee': 'Full Time',
    'Full Time/ Employee ': 'Full Time',
    'Full Time / > Employee': 'Full Time',
    'Job Type Full Time Employee': 'Full Time',
    'Full Time / Employee': 'Full Time',
    'Per Diem Employee': 'Per Diem',
    'Part Time Temporary/Contract/Project': 'Part Time',
    'Full Time Intern': 'Full Time',
    'Part Time Intern': 'Part Time',
    'Job Type Full Time Temporary/Contract/Project': 'Full Time',
    'Job Type Full Time': 'Full Time',
    'Part Time / Employee': 'Part Time',
    'Job Type Part Time Employee': 'Part Time',
    'Full Time  Employee': 'Full Time',
    'Full Time / Temporary/Contract/Project': 'Full Time'
})
df['job_type'] = df['job_type'].str.replace('Employee', '')
df['job_type'] = df['job_type'].str.replace('Per Diem', 'Part Time')
df['job_type'] = df['job_type'].str.replace('Part Time Seasonal', 'Part Time')
df['job_type'] = df['job_type'].str.replace('Part Time ', 'Part Time')
df['job_type'] = df['job_type'].str.replace('Full Time Temporary/Contract/Project', 'Temporary/Contract')
df['job_type'] = df['job_type'].str.replace('Part Time/ Temporary/Contract/Project', 'Temporary/Contract')
df['job_type'] = df['job_type'].str.replace('Full Time/', 'Full Time')
df['job_type'] = df['job_type'].str.replace('Full Time\xa0', 'Full Time')
df['job_type'] = df['job_type'].str.replace('Full Time\xa0 ', 'Full Time')
df['job_type'] = df['job_type'].str.replace('Part Time\xa0', 'Part Time')
df['job_type'] = df['job_type'].str.replace('Exempt', 'Temporary/Contract')
df['job_type'] = df['job_type'].str.replace('Job Type', 'Temporary/Contract')
df['job_type'] = df['job_type'].str.replace('Temporary/Contract/Project', 'Temporary/Contract')
df['job_type'] = df['job_type'].str.replace('Temporary/Contract ', 'Temporary/Contract')
df['job_type'] = df['job_type'].str.replace('Full Time ', 'Full Time')



```

```python
df['job_type'].value_counts()
```

```python
# Function to fill missing values based on job description
def fill_job_type(row):
    if pd.isna(row['job_type']):
        title = row['job_title'].lower()
        if "part time" in title:
            return "Part Time"
        elif "temporary" in title or "contract" in title:
            return "Temporary/Contract"
        elif "full time" in title:
            return "Full Time"
        else:
            return "Temporary/Contract"  # or any other default value you prefer
    else:
        return row['job_type']

# Apply the function to fill missing values in the 'job_type' column
df['job_type'] = df.apply(fill_job_type, axis=1)
```

```python
df['job_type'].value_counts()
```

### working on Location

```python
# Assume that location cannot be longer than 8 words (helps to filter out job descriptions that accidentially got here)
df['location'] = df.location.apply(str.strip)
df = df[df.location.apply(lambda x: len(x.split())) < 8]

# Get rid of excess words and symbols
df['location'] = (df.location.apply(lambda x: x.replace('Address ', ''))
                                 .apply(lambda x: x.replace(' \xa0 ', ' '))
                                 .apply(lambda x: x.replace(' \xa0', ' '))
                                 .apply(lambda x: x.replace('\xa0', ' '))
                                 .apply(lambda x: x.replace(' , ', ', ').strip()))


list(df.sample(30).location.unique())

formats = ['city, state, state postcode']
templates = ["^[A-Z][a-z\.\-'/ ]+( ?[A-Z][a-z\.\-'/]*){0,3}, [A-Z]{2} [0-9]{4,5}$"]
counts = []
# 3
df.location = df[df.location.str.match(templates[0])].location
```
