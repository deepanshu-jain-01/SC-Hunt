# DB URLs

```python
db_url = { 'Full':"https://sc-hunt-1-default-rtdb.firebaseio.com/",
           'Part':"https://sc-hunt-2-default-rtdb.firebaseio.com/",
           'Temp':"https://sc-hunt-3-bddff-default-rtdb.firebaseio.com/"}
```

```python
for i in range(df.shape[0]):
  value = json.dumps(df.loc[i].to_dict())
  data = json.loads(value)
  if('Full' in data['job_type']):
    db = db_url['Full']
  if('Part' in data['job_type']):
    db = db_url['Part']
  if('Temp' in data['job_type']):
    db = db_url['Temp']

  response = requests.put(db+str(i)+".json",value)
```
